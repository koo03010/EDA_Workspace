```python
req = driver.page_source
soup = BeautifulSoup(req, 'html.parser')
```

```python
location = soup.select('#mCSB_3_container > ul > li')
```

```python
# 가게이름 : #mCSB_3_container > ul > li:nth-child(6) > strong
# 주소 : ##mCSB_3_container > ul > li:nth-child(6) > p

store_list = []

for sto in tqdm_notebook(location):
    store_name = sto.find('strong').text
    store_addr = sto.find('p').text.replace('1522-3232','')
    store_gu = store_addr.split(' ')[1] # 'NoneType' object is not callable
    lat = sto["data-lat"]
    lng = sto["data-long"]

    store_list.append({
        'name' : store_name,
        'address' : store_addr,
        'gu' : store_gu,
        'lat' : lat,
        'lng' : lng
    })
```

```python
starbucks_gu = list(starbucks_df["gu"].unique())
seoul_gu = [str('서울 ') + gu for gu in starbucks_gu]
```

```python
ediya_url = 'https://ediya.com/contents/find_store.html'
driver = webdriver.Chrome('../driver/chromedriver.exe') 
driver.get(ediya_url) 

addr_click = driver.find_element(By.CSS_SELECTOR, '#contentWrap > div.contents > div > div.store_search_pop > ul > li:nth-child(2) > a')
addr_click.click()

keyword_search = driver.find_element(By.CSS_SELECTOR, '#keyword')
search_button = driver.find_element(By.CSS_SELECTOR, '#keyword_div > form > button')

ediya_list = []

for gu in tqdm_notebook(seoul_gu):
    # '커피랩'이라는 default 글자를 지워준다
    keyword_search.clear()

    # 검색어 seoul_gu에서 받아와서 입력
    keyword_search.send_keys(gu)

    # 검색버튼 클릭
    search_button.click()

    req = driver.page_source
    soup = BeautifulSoup(req, 'html.parser')

    # 검색결과
    location_ediya = soup.select('#placesList > li')

    # 이름, 주소, 구 정보
    for item in location_ediya:
        store_name = item.select_one('dt').text
        store_addr = item.select_one('dd').text
        store_gu = store_addr.split(' ')[1]
        ediya_list.append({
            'storeName': store_name,
            'address': store_addr,
            'gu': store_gu
        })
```
